```
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

Hello World

```
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer="sgd", loss="mean_squared_error")
# A dense layer is a layer of connected neurons
xs = np.array([-1,0,1,2,3,4], dtype=float)
ys = np.array([-3,-1,1,3,5,7], dtype=float)
```

```
model.fit(xs, ys, epochs=1000)
#We are asking the model to fit the x to the y and create a mapping in 1000 loops.
# Each Epoch will try to fit the model, calculate the loss and the optimize the model
print("\n")
print(model.predict([10.0]))
```

We can see that in the first few intial epochs, the cost is really high, but after that it begins to reduce sufficently (which is favourable)
